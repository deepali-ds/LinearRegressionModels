### 6) Outlier Analysis 

Outlier analysis or Anomaly detection is the identification of rare items, events or observations which raise suspicions by differing significantly from the majority of the data. Here, we are trying to identify oultiers in the temp, humid and gas_scan columns. So, that these suspicions observations can be taken care of in outlier treatment. 

##### 6.1) Train Set

```python
from collections import Counter
```

```python
measurements = ['temp_0', 'temp_1', 'temp_2', 'temp_3',
       'temp_4', 'temp_5', 'temp_6', 'temp_7', 'temp_8', 'temp_9', 'humid_0',
       'humid_1', 'humid_2', 'humid_3', 'humid_4', 'humid_5', 'humid_6',
       'humid_7', 'humid_8', 'humid_9', 'gas_scan_0', 'gas_scan_1',
       'gas_scan_2', 'gas_scan_3', 'gas_scan_4', 'gas_scan_5', 'gas_scan_6',
       'gas_scan_7', 'gas_scan_8', 'gas_scan_9']
```

```python
important_features = measurements + ['gas_scan_cnt', 'experiment', 'sensor_node_id', 'encoded_specimen']
```

```python
def find_outliers(df, measurements=measurements, cutoff=1.5):
    # initialize counters, etc.
    total_outliers = 0
    count_outliers_2 = 0
    rows_with_outliers = set() # set of row indices with outliers
    rows_count = []            # list of row indices with outliers
                                    # allows double counting
    df_binary = pd.DataFrame(index = df.index)
    
    # find outliers separately for clean_air (0) and in_smoke (1)
    for code in [0,1]:
        # select rows for analysis
        df_filt = df[df['encoded_specimen']==code]   

        # loop through all features
        for meas in measurements:
            # print(meas, code)
            # define outliers
            q25, q75 = df_filt[meas].quantile([.25,.75])
            iqr = q75-q25
            upper = q75 + cutoff * iqr
            lower = q25 - cutoff * iqr

            low_outliers = df_filt[df_filt[meas] < lower]
            hi_outliers = df_filt[df_filt[meas] > upper]
            total_outliers += low_outliers.shape[0]
            total_outliers += hi_outliers.shape[0]
            indices = list(low_outliers.index) + list(hi_outliers.index)
            if code==0:
                df_binary[meas] = 0
            df_binary.loc[indices, meas] = 1

            if code==1:
                try:
                    count_outliers_2 += df_binary[meas].value_counts()[1]
                except:
                    print('no outliers')
            # print(f'total outliers: {low_outliers.shape[0] + hi_outliers.shape[0]}')
            
            
            rows_lo =set(low_outliers.index) 
            rows_hi =set(hi_outliers.index) 
            rows_with_outliers = rows_with_outliers.union(rows_lo, rows_hi)
            
            
            # add row indices to list -- allowing multiple counts
            rows_count.extend(list(rows_lo))
            rows_count.extend(list(rows_hi))
        
    print(f'cutoff: {cutoff}')
    
    print(f'total outliers (all cells): {total_outliers}')
    print(f'count outliers 2 : {count_outliers_2}')
    print(f'rows with outliers: {len(rows_with_outliers)}')
    
    return rows_with_outliers, rows_count, df_binary
```

```python
rows_with_outliers, rows_count, df_binary = find_outliers(df=df_train, measurements=measurements, cutoff=1.5)
```

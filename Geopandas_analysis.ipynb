```python
import rasterio 
from matplotlib import pyplot as plt
from rasterio.plot import show
from rasterio.plot import show_hist
data = rasterio.open('/content/1040010055A5E300-visual.tif')
```

**Dataset attributes:
Properties of the raster data stored in the example GeoTIFF can be accessed through attributes of the opened dataset object. Dataset objects have bands and this example has a band count of 3.**

```python
data.count
```

**A dataset band is an array of values representing the partial distribution of a single variable in 2-dimensional (2D) space. All band arrays of a dataset have the same number of rows and columns.**

```python
show(data)
```

```python
print('data width is :',data.width,'\n data height is :', data.height)
```

```python
{i: dtype for i, dtype in zip(data.indexes, data.dtypes)}
```

```python
show_hist(
    data, bins=50, lw=0.0, stacked=False, alpha=0.3,
    histtype='stepfilled', title="Histogram")
```

```python
fig, (axr, axg, axb) = plt.subplots(1,3, figsize=(21,7))
show((data, 1), ax=axr, cmap='Reds', title='red channel')
show((data, 2), ax=axg, cmap='Greens', title='green channel')
show((data, 3), ax=axb, cmap='Blues', title='blue channel')
plt.show()
```
